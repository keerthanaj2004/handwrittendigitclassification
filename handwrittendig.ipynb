```python
import tensorflow as tf 
from tensorflow import keras
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
plt.matshow(x_train[0])
```

```python
y_train[0]
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train_new=x_train.reshape(len(x_train),(28*28))
x_test_new=x_test.reshape(len(x_test),(28*28))
```

```python
x_train.shape
```

```python
x_test_new[0]
```

```python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train_new,y_train,epochs=5)
```

```python
plt.matshow(x_test[3])
```

```python
ypred=model.predict(x_test_new)
```

```python
y_pred_labels=[np.argmax(i) for i in ypred]
y_pred_labels
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#adding hidden layers
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train_new,y_train,epochs=5)
```

```python
model.evaluate(x_test_new,y_test)
```

```python
#if u dont want to mess around with flatten array
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train,y_train,epochs=5)
```
